```python
import numpy as np
import pandas as pd
```

```python
# Load the dataset

df = pd.read_csv('datasets/all_talks_train.tsv', sep='\t')
```

```python
df.head()
```

```python
languages = df.columns[1:]

print(languages.tolist())

print(len(languages.tolist()))
```

```python
# Count __NULL__ entries per language

null_counts = {lang: int((df[lang] == '__NULL__').sum()) for lang in languages}
print(null_counts)
```

```python
# Reverse sort languages by number of null entries

sorted_null_counts = dict(sorted(null_counts.items(), key=lambda item: item[1], reverse=True))
print(sorted_null_counts)
```

```python
selected_languages = ['bn', 'be', 'kk']

selected_columns = df[selected_languages]

selected_columns.head()
```

```python
cleaned_df = selected_columns[~selected_columns.apply(lambda x: x.str.contains('NULL')).any(axis=1)]

cleaned_df
```
